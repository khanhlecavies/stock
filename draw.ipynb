```python
import json
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Đọc dữ liệu từ tệp JSON
with open("data.json", "r") as file:
    data = json.load(file)

# Lấy giá mở cửa từ dữ liệu JSON
open_prices = data["o"]

# Vẽ biểu đồ histogram
plt.hist(open_prices, bins=30, density=True, alpha=0.6, color="b")

# Tính toán phân phối chuẩn
mu, sigma = np.mean(open_prices), np.std(open_prices)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, "k", linewidth=2)
title = "Fit results: mu = %.2f,  sigma = %.2f" % (mu, sigma)
plt.title(title)

plt.show()
```
